# inner Join

```python
import pandas as pd
df_movies= pd.read_csv("../day12/IMDb movies.csv/IMDb movies.csv", low_memory = False)
```

```python
df_movies
```

```python
df_rating= pd.read_csv("../day12/IMDb ratings.csv/IMDb ratings.csv")
df_rating
```

```python
df_movies_inner_join = df_movies.merge(df_rating ,on =  'imdb_title_id', how= 'inner')
#pd.merge(df_movies, df_rating ,on =  'imdb_title_id') #does the same thing by default it's inner
df_movies_inner_join 
```

# full join `outer join`

```python
df_movies_outer_join = pd.merge(df_movies,df_rating , on = 'imdb_title_id', how = 'outer')
df_movies_outer_join
```
