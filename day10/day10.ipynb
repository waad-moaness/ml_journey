```python
import pandas as pd 
df= pd.read_csv('https://raw.githubusercontent.com/thepycoach/python-for-data-science/main/07.Project%202%20-%20Data%20Visualization/population_total.csv')
```

```python
df
```

```python
# drop null values
df.dropna(inplace = True )
```

```python
df.info()
```

```python
# reshape the df so it make more sense 
df_pivot= df.pivot_table(index= 'year' , columns= 'country' , values='population', aggfunc= 'sum')
```

```python
df_pivot.dropna(inplace = True )
df_pivot
```

```python
# select specific countries
df_pivot= df_pivot[['United States', 'Egypt', 'China', 
                          'Indonesia', 'Brazil']]
```

```python
df_pivot
```

# Lineplot

```python
df_pivot.plot(kind = 'line' , xlabel = 'years', ylabel= 'population' , figsize= (8,4), title= 'population over the years')
```

```python
# rename a column 
df_pivot_2015 = df_pivot[df_pivot.index.isin([2015])]
df_pivot_2015
```

```python
# reverse columns and rows 
df_pivot_2015= df_pivot_2015.T
df_pivot_2015
```

# barplot

```python
df_pivot_2015.plot(kind = "bar", color= 'purple')
```

# pie chart

```python
# first change the year column to a string so we ca use it as a y label 
df_pivot_2015.rename(columns= {2015: "2015"} , inplace = True)
```

```python
df_pivot_2015.plot(kind = 'pie', y= "2015")
```
