```python
# query method
df.query("age> 30 and nationality == 'Croatia'")
# boolean slicing
# df[df.age > 30]
```

```python
df.dob
# type object to be able tp query it i need to change it to datetime64 type 
df.dob = df.dob.astype('datetime64[ns]')
df.dob
```

```python
df.query("dob.dt.year > 1990")
```

to make opertions to more than one column we must use the axis =1 parameter in the apply method 

```python
# convert long name to uppercase()
df.long_name.apply(lambda x: x.upper())
# x represent one element in each row at a time
df.long_name.str.upper() # does the same thing i use the str attribute to access the string methods 
```

```python
# get year from dob
df.dob.apply(lambda x:x.year)
# or use dt attributes
df.dob.dt.year
```

```python
# calculate bmi using lambda we need to specify the axis =1 cuz we are using more than one column 
df.apply(lambda x: x['weight_kg']/ ((x['height_cm']/100)** 2) , axis =1)
```

to copy data frames we use copy() 

```python
# deep parameter specify it changes in the original will be reflected in the copy or not (default = True => don't reflect)
df_copy = df.copy()
df_copy = df.copy(deep = False )# reflects ( we call it shallow copy)
df_new = df # is also a shallow copy 
```

```python
df_table.pivot_table(index='Gender', values=df_table.select_dtypes(include='number').columns, aggfunc='sum').round(1)
# in older pandas versions (before ~v1.5):

# .pivot_table(aggfunc='sum') automatically ignored non-numeric columns, like datetime.

# It just summed the numeric ones — silently.

# 🟢 So the tutorial works because pandas silently skipped datetime columns like Date and Time.
# so we use this for it to work {values=df_table.select_dtypes(include='number').columns} => Only uses numeric columns
```

```python
# make a pivot tables that says how much male and female spend in each category
df_table.pivot_table(index = 'Gender', columns= 'Product line', values= 'Total', aggfunc='sum')
```

| Parameter  | What it does                            |
| ---------- | --------------------------------------- |
| `index=`   | Becomes row labels (group by this)      |
| `values=`  | Which columns to aggregate (not pivot)  |
| `columns=` | Creates new columns (used for pivoting) |
