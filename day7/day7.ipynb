```python
# create a new column based on the screan size 
import numpy as np # so we can use where() 

df['Screen_size'] = np.where(df.Inches > 13 , 'Big Screen', 'Small Screen')
```

```python
# create a new column based on multiple condition select()
conditions=[
    df.Price_euros >=  3000,
    ( (df.Price_euros >= 2000) &  (df.Price_euros < 3000)),
     df.Price_euros < 2000
]
values = ['expensive' , 'mid', 'cheap']
```

```python

df['price_tiers'] = np.select(conditions, values, default='unknown')
df
```

```python
# filter multiple conditions using isin()
df[(df.Company.isin(["Apple","HP"])) & (df.TypeName.isin(["Ultrabook","Notebook"]))].value_counts(['Company', 'TypeName'])
```

```python
# sort df by company and price
df2 = df.sort_values(['Company', 'Price_euros'])
df2.Company.value_counts()
df2[df2.duplicated("Company", keep = "first")] #show duplecated observations execluding first dup 
# keep last returns all dups execluding last dup , keep = False return all dups in the data
# to show data with no dups use the ~
df2[~df2.duplicated("Company", keep = "first")] [["Company","Price_euros"]]
```

```python
# if we check now we will see no dups
df2[~df2.duplicated("Company", keep = "first")]["Company"].value_counts()
```
