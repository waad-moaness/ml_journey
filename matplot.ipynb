```python
lang = ['python', 'c++' , 'go' , 'js' , 'c' ]
```

```python
values = [60,74 ,57,6,66]
```

```python
plt.pie(values , labels= lang, explode= [0,0.3,0,0.7,0]) 
plt.show()
```

explode drages the ones with values out alittle bit  the bigger the number the farther it get

```python
plt.pie(values , labels= lang, autopct= '%.2f%%') 
plt.show()
```

```python
plt.pie(values , labels= lang, autopct= '%.3f%%',pctdistance= 0.7) 
plt.show()
```

```python
plt.pie(values , labels= lang, autopct= '%.3f%%',pctdistance= 0.7, startangle= 90) 
plt.show()
```
